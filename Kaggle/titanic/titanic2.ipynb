テストデータの'PassengerId',　'Pclass', 'Sex', 'Age', 'Fare'の列を利用

訓練データは用いない　　

AgeとFareの欠損値は平均で埋める　　

Sexはfemaleを0、maleを1に変換

```python
import pandas as pd

test = pd.read_csv('test.csv')
test.head()
```

```python
test = test.drop(['Name', 'Ticket', 'Cabin', 'Embarked'], axis=1)
test.head()
```

```python
test["Sex"] = test["Sex"].replace(["female", "male"], [0, 1])
test.isnull().sum()
```

```python
test.head()
```

```python
test["Age"] = test["Age"].fillna(test["Age"].mean())
test['Fare'] = test['Fare'].fillna(test['Fare'].mean())
test.isnull().sum()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np

sc = StandardScaler()
test_sc = sc.fit_transform(test)

kmeans = KMeans(n_clusters=2, random_state=0)
clusters = kmeans.fit(test_sc)
test["cluster"] = clusters.labels_
test['Survived'] = np.where(test['cluster']==1, 0, 1)
test.head()
```

```python
test.groupby('Pclass')['Survived'].mean()
```

```python
pred2 = test[['PassengerId', 'Survived']]
pred2.to_csv('pred2.csv', index=False)
pred2.head()
```
