```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dt = pd.read_csv('train.csv')
dt.head(), dt.shape
```

```python
dt['Destination'].value_counts()
```

```python
dt['HomePlanet'].value_counts()
```

```python
dt.drop(['Cabin', 'Name'], axis=1, inplace=True)
```

```python
dt.head()
```

```python
dt['HomePlanet'] = dt['HomePlanet'].replace(['Earth', 'Europa', 'Mars'], [0, 1, 2])
dt.head()
```

```python
dt['Destination'] = dt['Destination'].replace(['TRAPPIST-1e', '55 Cancri e', 'PSO J318.5-22'], [0, 1, 2])
dt.head()
```

```python
dt['VIP'] = dt['VIP'].replace([True, False], [1, 0])
dt.head()
```

```python
dt['CryoSleep'] = dt['CryoSleep'].replace([True, False], [1, 0])
```

```python
dt['Transported'] = dt['Transported'].replace([True, False], [1, 0])
dt.head()
```

```python
dt.isnull().sum()
```

```python
dt["HomePlanet"] = dt["HomePlanet"].fillna(dt["HomePlanet"].mean())
dt["CryoSleep"] = dt["CryoSleep"].fillna(dt["CryoSleep"].mean())
dt["Destination"] = dt["Destination"].fillna(dt["Destination"].mean())
dt["Age"] = dt["Age"].fillna(dt["Age"].mean())
dt["VIP"] = dt["VIP"].fillna(dt["VIP"].mean())
dt["RoomService"] = dt["RoomService"].fillna(dt["RoomService"].mean())
dt["FoodCourt"] = dt["FoodCourt"].fillna(dt["FoodCourt"].mean())
dt["ShoppingMall"] = dt["ShoppingMall"].fillna(dt["ShoppingMall"].mean())
dt["Spa"] = dt["Spa"].fillna(dt["Spa"].mean())
dt["VRDeck"] = dt["VRDeck"].fillna(dt["VRDeck"].mean())
```

```python
dt.isnull().sum(), dt.shape
```

```python
dt.head()
```

```python
test = pd.read_csv('test.csv')
test.drop(['Cabin', 'Name'], axis=1, inplace=True)
test['HomePlanet'] = test['HomePlanet'].replace(['Earth', 'Europa', 'Mars'], [0, 1, 2])
test['Destination'] = test['Destination'].replace(['TRAPPIST-1e', '55 Cancri e', 'PSO J318.5-22'], [0, 1, 2])
test['VIP'] = test['VIP'].replace([True, False], [1, 0])
test['CryoSleep'] = test['CryoSleep'].replace([True, False], [1, 0])
test.isnull().sum()
```

```python
test["HomePlanet"] = test["HomePlanet"].fillna(test["HomePlanet"].mean())
test["CryoSleep"] = test["CryoSleep"].fillna(test["CryoSleep"].mean())
test["Destination"] = test["Destination"].fillna(test["Destination"].mean())
test["Age"] = test["Age"].fillna(test["Age"].mean())
test["VIP"] = test["VIP"].fillna(test["VIP"].mean())
test["RoomService"] = test["RoomService"].fillna(test["RoomService"].mean())
test["FoodCourt"] = test["FoodCourt"].fillna(test["FoodCourt"].mean())
test["ShoppingMall"] = test["ShoppingMall"].fillna(test["ShoppingMall"].mean())
test["Spa"] = test["Spa"].fillna(test["Spa"].mean())
test["VRDeck"] = test["VRDeck"].fillna(test["VRDeck"].mean())
```

```python
test.isnull().sum()
```

```python
from sklearn import svm
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.preprocessing import StandardScaler

# StandardScalerのインスタンスを作成する
scaler = StandardScaler()

# 学習用のデータを用意する
X = dt[['PassengerId', 'HomePlanet', 'CryoSleep', 'Destination', 'Age', 'VIP', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']] # 特徴量
y = dt['Transported'] # ラベル

# fitメソッドで学習用のデータの平均と分散を計算する
scaler.fit(X)

# transformメソッドでデータを標準化する
X_scaled = scaler.transform(X)

# SVCのインスタンスを作成する
clf = svm.SVC(gamma="scale")

# fitメソッドで学習させる
clf.fit(X_scaled, y)

# predictメソッドで予測する
test_scaled = scaler.transform(test) # テスト用のデータも標準化する
pred = clf.predict(test_scaled) # 予測する
df = pd.DataFrame(data=pred, columns=['Transported'])
rst = pd.merge(test, df, left_index=True, right_index=True)
rst.head()
```

```python
pred = rst[['PassengerId', 'Transported']]
pred.replace([1, 0], [True, False], inplace=True)
pred.to_csv('pred.csv', index=False)
pred.head()
```
